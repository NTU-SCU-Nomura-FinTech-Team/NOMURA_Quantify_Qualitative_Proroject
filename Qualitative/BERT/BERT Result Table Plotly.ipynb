```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Result_Table = pd.read_csv('Result_Table.csv')
# Result_Table = pd.read_csv('Result_Table_no_shift.csv')
Result_Table
```

```python
from plotly.graph_objs import Scatter, Layout
import plotly.graph_objs as go
import plotly 


#plotly.io.renderers.default = "svg" 

#  Jupyter 
#plotly.offline.init_notebook_mode(connected=True)

df = Result_Table.copy()

# 
fig = go.Figure()

# 
fig.add_trace(go.Scatter(x=df['Date'], y=df['Change'], mode='markers', name='SP500 change %'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Shift'], mode='markers', name='Shift change %'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Label'], mode='markers', name='Label'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Predict'], mode='markers', name='Predict'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['Probability'], mode='markers', name='Probability'))
fig.add_trace(go.Scatter(x=df['Date'], y=0.5*np.ones(len(df['Date'])), mode='lines', name='Split 0.5'))

# 
fig.update_layout(title_text='Predict Result', title_x=0.5)


# 
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1M", step="month", stepmode="backward"),
            dict(count=6, label="6M", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1Y", step="year", stepmode="backward"),
            dict(count=2, label="2Y", step="year", stepmode="backward"),
            dict(label="All",step="all")
        ])
    )
)
# 
#fig.show("svg")

# 
fig.show()
#plotly.offline.iplot(fig)
```
